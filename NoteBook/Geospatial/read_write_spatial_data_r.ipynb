## Vector Data

Vector data is the most widely accepted and accurate method for representing geospatial data in various fields, including Geographic Information Systems (GIS) and computer graphics. This technique utilizes points, lines, and polygons to illustrate spatial information with precision and clarity. Points signify a single location on a map, while lines connect two or more points, and polygons are closed shapes that represent enclosed areas. Each shape is defined by its geometric coordinates, which consist of x, y, and sometimes z values that describe the location of each point, line vertex, or polygon corner. Vector data enables easy manipulation and analysis of geospatial information, making it an essential tool for professionals who seek a confident and reliable representation of spatial data.

Here's a breakdown of the main elements of vector data:

1.  **Points**: A point vector is an essential component of geospatial data that confidently identifies a specific location in space by its precise x and y coordinates on a coordinate system. Points serve as the basis for various spatial applications, including mapping and Geographic Information Systems (GIS), to depict features such as cities, landmarks, and sampling locations. Also, points are the foundation for more intricate geospatial data, such as lines and polygons. Therefore, having a firm grasp of the concept of points and their role in geospatial data is integral to and will aid in creating accurate and reliable geospatial data.

2.  **Lines (Polylines)**: A line vector is a crucial element used to represent a variety of linear features such as roads, rivers, and boundaries. These lines are composed of a series of connected points that can be straight or curved, depending on the feature they represent. The shape of these lines is defined by multiple vertices that determine each segment's direction, angle, and length. These vertices act as anchor points that connect the individual points and create a seamless line that accurately captures the shape and path of the feature it represents.

3.  **Polygons**: Polygons are two-dimensional closed shapes that are formed by connecting a series of lines to enclose an area. They are commonly used in cartography and geographic information systems (GIS) to represent features such as countries, lakes, forests, and other spatially bounded regions. By defining the shape and boundary of an area, polygons provide a useful tool for analyzing and visualizing spatial data. They can also be used in various fields, such as computer graphics, engineering, and architecture, to represent complex shapes and structures.

![alt text](http://drive.google.com/uc?export=view&id=1tmGnqvHRbYqG3BYjPkn4cAhp3UFM2eEG)

Vector data provides an accurate and detailed representation of geographic features. In contrast to raster data, another commonly used data format in GIS, vector data allows for precise measurements, editing, and analysis of geographic features. It is beneficial for storing data with well-defined boundaries and discrete features and can be easily scaled to accommodate larger data sets without sacrificing accuracy or efficiency. With its high precision and flexibility, vector data is essential for anyone working with geographic information

Some common vector data formats include Shapefile (`.shp`), GeoJSON (`.geojson`), and Keyhole Markup Language (`.kml`). These formats can store various attributes along with the spatial coordinates, making it possible to attach additional information (e.g., population data for cities or elevation data for terrain) to each feature represented in the vector dataset.

The **shapefile**, a interchangeably data format which are regulated by [ESRI](https://www.esri.com/en-us/home). It is one of the most common form of geospatial vector data used in GIS software and analyses. Three unique files are required for a shapefile, including: `shx` (shape index format; this tags the shapefile with a position, so users can move it forward and backward among layers, a `.shp` (shape format, which stores geometric data, and a `.dbf` (attribute format file; which holds attributes (information) for the shapes in the file).

The **GeoJSON** is a file format specifically designed to represent geographical data, making storing, sharing, and analyzing location-based information easier. It uses a simple and widely recognized syntax based on **JavaScript Object Notation (JSON)** to represent geographic features such as points, lines, polygons, and their properties. This allows for creating detailed maps, visualizations, and other geographic applications. Using GeoJSON, developers, and analysts can easily access and manipulate geographical data, making it an essential tool for many industries.

**Keyhole Markup Language (KML)** is a file format that enables creating and sharing geospatial data in a visually compelling 3D environment. KML files contain instructions that dictate how geographic data should be displayed on virtual globes, such as Google Earth and Google Maps. Initially created by Keyhole, Inc., KML was designed to provide a simple and efficient way to represent complex geographical data. Later, in 2004, Google acquired Keyhole, Inc. and integrated KML into their virtual globe applications. The versatility of KML files lies in their ability to incorporate a wide range of data types, including maps, images, 3D models, videos, and even sound recordings. This has made KML an indispensable tool for visualizing geographical data in an informative and engaging way. Whether you're a scientist, engineer, educator, or just a curious individual, KML files can help you explore and share the world around you in a whole new way.

## Raster data

Raster data is the go-to spatial data for representing spatial information in various contexts. Instead of vector data, which uses points, lines, and polygons to represent spatial features, raster data is structured into a grid-like format, where each cell or pixel in the grid contains a value representing a specific attribute or measurement at a specific location. This structured format provides a highly detailed and nuanced representation of the landscape, with each pixel offering valuable information about the environmental, topographical, or other characteristics of the area it represents. Using raster data, analysts and decision-makers can gain a complete and nuanced understanding of the landscape, enabling them to make better-informed decisions about managing and utilizing it.

Key characteristics of raster data include:

1.  **Pixel-based representation**: Raster data divides the geographic area into a regular grid of pixels, where each pixel corresponds to a specific location on the Earth's surface. Each pixel stores a value representing a particular attribute, such as elevation, temperature, land use, satellite imagery, or any other continuous data.

2.  **Resolution**: The resolution of raster data refers to the size of each pixel, which determines the level of detail in the data. Higher resolution means smaller pixel sizes and more detail, while lower resolution means larger pixel sizes and less detail.

3.  **Cell values**: The cell values in raster data can represent a wide range of information, such as elevation above sea level, temperature readings, land cover categories, or satellite sensor measurements.

4.  **Coverage and extent**: Raster datasets have a defined coverage area and extent, meaning they represent a specific geographic region with boundaries.

Common examples of raster data include digital elevation models (DEMs) used for terrain representation, satellite imagery, aerial photographs, climate data, and remotely sensed data from various sensors.

Raster data has several advantages, including its ability to efficiently store and process large continuous datasets and perform spatial analysis like interpolation, proximity analysis, and slope calculations. However, it may not be as accurate or suitable for representing discrete features with well-defined boundaries compared to vector data.

Raster data formats often include GeoTIFF (.tif), JPEG (.jpg), PNG (.png), and many other specialized formats depending on the type of data being represented and the intended use.

Unlike vector data, a raster data consists of cells or pixels organized into rows and columns as a matrix where each cell contains a value representing geographical feature on the earth. The size of the area on the surface that each pixel covers is known as the spatial resolution of the image. For instance, an image that has a 1 m spatial resolution means that each pixel in the image represents a 1 m x 1 m area. There are two types of raster data: continuous and discrete. An example of discrete raster data is Land use raster. Data types are flexible, including discrete and categorical data, such as soil or land-use maps, or continuous data as in digital elevation models, precipitation gradient maps, or pollutant concentration maps, etc.

![alt text](http://drive.google.com/uc?export=view&id=1tl830Uz5EpCqzlbx8dZueH9dLXt2ZB99)


In GIS and remote sensing applications, there are generally two types of raster data sets: single-band and multiple-band rasters. Single-band rasters measure only one characteristic of a spatial area and contain a single layer of cell values representing that characteristic. For instance, a digital elevation model (DEM) is an example of a single-band raster, where each cell contains only one value representing the elevation of the earth's surface.

On the other hand, most satellite imagery has multiple bands. A multiple-band raster contains multiple spatially corresponding matrices of cell values representing the same spatial area. Typically, these bands represent different ranges of the electromagnetic spectrum.

#### Single band raster

A single band raster, also known as a single-band image or grayscale image, is a type of raster data that contains only one band or layer of information. In a single-band raster, each pixel is represented by a single value, typically indicating a certain property or characteristic at that location. The value of each pixel in the raster corresponds to the magnitude of the property being measured at that specific location. Digital Elevation Models (DEMs) represent the elevation or height values of the Earth's surface at each pixel location. DEMs are often used in terrain analysis, hydrology, and 3D visualization.

#### Multi-bands raster

A multiband raster is a type of raster data composed of multiple layers or bands of information. Each band captures data on a specific property or characteristic at each pixel location on the Earth's surface. These properties or characteristics can include temperature, reflectance, or vegetation index. Combining information from multiple bands can create a more comprehensive picture of the Earth's surface.

Multiband rasters are commonly used in remote sensing, GIS, and image processing applications. In remote sensing, multiband rasters are used to capture data from various sensors, such as satellites or aircraft, to analyze and monitor changes on the Earth's surface over time. Multiband rasters create maps in GIS applications and analyze spatial relationships between features. In image processing, multiband rasters enhance and manipulate images for various purposes, such as identifying objects or features in an image.

Multiband rasters are essential for capturing and analyzing complex data that cannot be represented in a single band. For example, a single band may not be able to capture both vegetation and water in an image. Multiple bands can capture vegetation and water with a multiband raster, allowing for more accurate and detailed image analysis.


For loading multi-bands image in R, we will use a sub-set of Landsat 9 multispectral image covering a rice grwoing area of Bangladesh. This image was downloaded from [USGS Earth Explore](http://earthexplorer.usgs.gov/).

You can read single band image one by one with `raster()` function of **raster** package.

```python
%%R
landsat9 = '/content/drive/MyDrive/R_Website/Geospatial/Data/Bangladesh/Raster/Landsat/LC09_L2SP_138043_20221216_20221223_02_T1'
```

```python
%%R
# read band 2 (Blue band)
b2 <- raster::raster(paste0(landsat9, '/LC09_L2SP_138043_20221216_20221223_02_T1_SR_B2.TIF'))
# read band 3 (Green band)
b3 <- raster::raster(paste0(landsat9, '/LC09_L2SP_138043_20221216_20221223_02_T1_SR_B3.TIF'))
# read band 4 (Red band)
b4 <- raster::raster(paste0(landsat9, '/LC09_L2SP_138043_20221216_20221223_02_T1_SR_B4.TIF'))
# read band 5 (NIR band)
b5 <- raster::raster(paste0(landsat9, '/LC09_L2SP_138043_20221216_20221223_02_T1_SR_B5.TIF'))
```

Now, you can use `stack()` or `brick()` function of raster package to create multi-band raster stack

```python
%%R
multiband_01<-raster::stack(b2,b3,b4,b5)
plot(multiband_01)
```

OR you can create a list of raster layers to use (for example bands 2 to 5) first then use stack() function to raster list read all several raster

```python
%%R
landsat9 = '/content/drive/MyDrive/R_Website/Geospatial/Data/Bangladesh/Raster/Landsat/LC09_L2SP_138043_20221216_20221223_02_T1'
bandList <- paste0(landsat9,"/LC09_L2SP_138043_20221216_20221223_02_T1_SR_B", 2:5, ".TIF")
bandList
```

Can can create a multi-band raster using `rast()` function of **terra** package

```python
%%R
multiband_02 <- terra::rast(bandList)
multiband_02
```

We can plot `SpatRaster` of terra pacakge using `geom_spatraster()` of  tidyterra package

```python
%%R -w 700 -h 700 -u px

ggplot() +
  geom_spatraster(data = multiband_02) +
  facet_wrap(~lyr, ncol = 2) +
  scale_fill_whitebox_c(
    palette = "muted",
    n.breaks = 12,
    guide = guide_legend(reverse = TRUE)
  ) +
  labs(
    fill = "",
   title = "Four Bands of Landsat 9",
  )
```

Can can also create a multi-band raster using `read_stars()` function of **stars** package:

```python
%%R -w 600 -h 600 -u px
multiband_03 <- stars::read_stars(bandList)
multiband_03
```

We can  visualize false color composite using `plot RGB()` function:

```python
%%R -w 500 -h 500 -u px
# False color composite NIR:Red:Green
plotRGB(multiband_02, r = 4, g = 3, b = 2, stretch="lin")
```

## GeoJson data

GeoJSON is a format for encoding geographic data structures using JavaScript Object Notation (JSON). It is commonly used for representing geographical features and their attributes such as points, lines, polygons, and more. GeoJSON is widely used in web mapping applications and geographic information systems (GIS) due to its simplicity, easy integration with web technologies, and human-readability.

Here's an example of a simple GeoJSON representation for a point feature:



```python
{
  "type": "Feature",
  "geometry": {
    "type": "Point",
    "coordinates": [longitude, latitude]
  },
  "properties": {
    "name": "Sample Point",
    "description": "This is an example of a GeoJSON point feature."
  }
}
```

Here's an explanation of the key components:

-   **`type`**: Specifies the type of the GeoJSON object. It can be "Feature," "FeatureCollection," "Point," "LineString," "Polygon," and others.

-   **`geometry`**: Contains the geometric data for the feature, such as the coordinates for a point, the array of coordinates for a line or polygon, etc.

-   **`coordinates`**: The array of coordinates in the format **`[longitude, latitude]`** for points and **`[ [lon1, lat1], [lon2, lat2], ..., [lonN, latN] ]`** for lines and polygons.

-   **`properties`**: An object that holds additional properties or attributes associated with the feature.

## KML

Keyhole Markup Language (KML) is an XML-based file format used to represent geographic data and features in 3D Earth browsers, such as Google Earth, Google Maps, and other mapping applications. KML was developed by Keyhole, Inc., which was acquired by Google in 2004, and it became an open standard in 2008.

KML files are structured as XML documents and are used to describe and display various types of geographical data, including points, lines, polygons, images, and overlays. KML allows you to define the geometry, attributes, and styles of these features. It is particularly useful for creating custom geographic content that can be displayed on digital maps and shared easily.

```python
<?xml version="1.0" encoding="UTF-8"?>
<kml xmlns="http://www.opengis.net/kml/2.2">
  <Placemark>
    <name>Sample Point</name>
    <description>This is an example of a KML point feature.</description>
    <Point>
      <coordinates>-122.4194,37.7749,0</coordinates>
    </Point>
  </Placemark>
</kml>
```
