## Introduction

The Landsat series of satellites is one of the most well-known and longest-running Earth observation programs, operated by NASA and the U.S. Geological Survey (USGS). The Landsat satellites have played a pivotal role in monitoring and studying the Earth's land surfaces and have provided valuable data for various applications, including agriculture, forestry, environmental monitoring, urban planning, and more.

Landsat satellites are equipped with sensors that capture images in various wavelengths of light, ranging from visible to infrared. These images are used for a wide range of applications, including land cover and land use monitoring, environmental assessment, natural resource management, urban planning, disaster response, and scientific research.

The Landsat satellites orbit the Earth in a sun-synchronous polar orbit, which means they pass over the same area at approximately the same local solar time on each orbit. This consistent viewing geometry allows for the collection of images under consistent lighting conditions, making it easier to compare changes in the landscape over time.

One of the significant features of Landsat images is their ability to capture multispectral data, which means they record information in different wavelength bands beyond what is visible to the human eye. This provides valuable insights into the composition and health of vegetation, the presence of water bodies, and various geological and environmental features.

**Wavelength and Resolution of Different Landsat Satellites**

Landsat satellites have varied in terms of the number of spectral bands (wavelengths) they observe and the spatial resolution of their imagery. As of my last knowledge update in September 2021, here is a summary of the wavelength ranges and spatial resolutions for different Landsat satellites:

1.  **Landsat 1-3 (Landsat 1 launched in 1972, Landsat 3 in 1978):**

    -   Spectral Bands: 4 visible and near-infrared bands (0.45 - 1.2 micrometers).

    -   Spatial Resolution: Approximately 79 meters for multispectral data, 185 meters for the panchromatic band.

2.  **Landsat 4-5 (Landsat 4 launched in 1982, Landsat 5 in 1984):**

    -   Spectral Bands: 7 bands, including visible, near-infrared, and thermal infrared (0.45 - 12.5 micrometers).

    -   Spatial Resolution: Approximately 30 meters for multispectral data, 120 meters for the thermal band.

3.  **Landsat 6 (Launched in 1993, but failed to reach orbit):**

    -   Landsat 6 was intended to carry similar sensors as Landsat 4-5 but did not become operational due to a launch failure.

4.  **Landsat 7 (Launched in 1999):**

    -   Spectral Bands: 8 bands, including visible, near-infrared, and thermal infrared (0.45 - 12.3 micrometers).

    -   Spatial Resolution: Approximately 30 meters for multispectral data, 60 meters for the panchromatic band, and 120 meters for the thermal band.

5.  **Landsat 8 (Launched in 2013):**

    -   Spectral Bands: 11 bands, including visible, near-infrared, shortwave infrared, and thermal infrared (0.43 - 12.5 micrometers).

    -   Spatial Resolution: Approximately 30 meters for multispectral data, 15 meters for the panchromatic band, and 100 meters for the thermal band.
    
6.  **Landsat 9 (Lunched in 2021)**

    -   Landsat 9 is the latest addition to the Landsat series of Earth observation satellites. It continues the legacy of its predecessors in providing high-quality satellite imagery and data for various applications, including environmental monitoring, land use planning, agriculture, forestry, and disaster response

**Landsat Image Processing levels**

Landsat images are distributed in different processing levels, each representing a specific stage of data processing and calibration. These levels provide varying levels of processing and correction, catering to different user needs and applications. As of my last knowledge update in September 2021, Landsat imagery was typically distributed in the following processing levels:

1.  Level-0 (L0): This is the raw, unprocessed data directly received from the satellite's sensors. It includes the digitized signals from the detectors and usually requires significant calibration and processing before it can be used effectively.

2.  Level-1 (L1) Terrain-Corrected (L1T): In this processing level, the Level-0 data is geometrically corrected to remove distortions caused by the satellite's orbit and Earth's curvature. The result is a radiometrically and geometrically corrected image, but without ground control points.

3.  Level-1 (L1) GeoTIFF (L1G): Similar to L1T, this processing level includes radiometric and geometric corrections, but with additional tie points or ground control points to improve the accuracy of the geolocation information. This allows for better alignment with existing maps and datasets.

4.  Level-2 (L2) Surface Reflectance: L2 processing involves atmospheric correction to remove the effects of atmospheric scattering and absorption, resulting in a more accurate representation of the surface reflectance values. This level is suitable for various land cover and land use applications and allows for meaningful comparisons of reflectance across time and space.

5.  Level-2 (L2) Surface Temperature: This level involves processing thermal infrared data to derive land surface temperatures. It's particularly useful for applications related to studying urban heat islands, hydrology, and environmental monitoring.

6.  Level-3 (L3): L3 products are derived from multiple scenes over a specific time period and are often used to create composite images that provide a broader view of changes in land cover, vegetation health, or other features over larger areas.

7.  Level-4 (L4): These products are derived from more complex analyses or modeling processes and often include data from multiple sources, not just Landsat. They can include things like land cover classifications, vegetation indices, and more advanced geospatial products.

It's important to note that processing levels can vary depending on the specific Landsat satellite in question (e.g., Landsat 5, 7, 8) and the data provider. Additionally, advancements in technology and processing techniques may have led to changes or refinements in these processing levels since my last update. For the most up-to-date information, I recommend visiting the official Landsat program websites maintained by the United States Geological Survey (USGS) or NASA.

### Read meta-data (MLT) file

**readMeta()** function of RStoolbox reads legacy versions of Landsat metadata files (\_MLT.txt) and where possible adds missing information (radiometric gain and offset, earth-sun distance).

The metadata for Landsat 8 is typically provided in a format known as the "Landsat Metadata File" (MTL file). Here is an overview of the key metadata contained in a Landsat 8 MTL file:

1.  **General Information:**

    -   **Scene ID:** A unique identifier for the Landsat scene.

    -   **Acquisition Date and Time:** The date and time when the image was acquired.

    -   **Spacecraft ID:** The identifier for the Landsat 8 satellite.

    -   **Sensor ID:** The identifier for the specific sensor (e.g., OLI/TIRS).

2.  **Geospatial Information:**

    -   **WRS Path and Row:** The Landsat Worldwide Reference System (WRS) path and row number that specify the scene's location on Earth.

    -   **Map Projection Information:** Details about the map projection used for the image, including the projection type, datum, and parameters.

3.  **Radiometric Information:**

    -   **Radiometric Calibration Factors:** Parameters related to sensor radiometric calibration, including gain and bias values for each spectral band.

    -   **Reflectance Conversion Factors:** Factors used to convert digital numbers (DN) to at-sensor reflectance values for each band.

4.  **Spectral Information:**

    -   **Spectral Bands:** Information about the spectral bands captured by the sensor, including central wavelengths and bandwidths.

5.  **Thermal Information (TIRS Bands):**

    -   **Thermal Band Calibration:** Parameters for calibrating the thermal infrared (TIRS) bands, including emissivity values and coefficients.

6.  **Sun and View Geometry:**

    -   **Solar Angle:** The solar zenith and azimuth angles at the time of image acquisition.

    -   **Sensor Elevation Angle:** The sensor's elevation angle during acquisition.

    -   **Sensor Azimuth Angle:** The sensor's azimuth angle during acquisition.

7.  **Image Quality and Cloud Cover:**

    -   **Image Quality:** Information about the image's quality, including the cloud and cloud shadow assessment.

    -   **Cloud Cover:** The estimated percentage of cloud cover in the scene.

8.  **Bounding Coordinates:**

    -   **Corner Coordinates:** The latitude and longitude coordinates of the four corners of the image extent.

9.  **Data Format Information:**

    -   **Data Format Version:** The version of the data format used for the Landsat scene.

10. **File Names and Paths:**

    -   **File Names:** The names of the image files associated with the scene, including the individual spectral bands.

These metadata are essential for properly interpreting and processing Landsat 8 imagery, including tasks such as atmospheric correction, geometric correction, and data analysis. Researchers and analysts often rely on this metadata to ensure accurate and consistent use of Landsat data for various applications, such as land cover mapping, environmental monitoring, and natural resource management.

### Pan Sharpening or Image Fusion

Pan sharpening, short for panchromatic sharpening, is a remote sensing and image processing technique used to enhance the spatial resolution of multispectral imagery by combining it with a higher-resolution panchromatic image. The goal of pan sharpening is to create a single high-resolution color image that retains the spectral information of the multispectral data while incorporating the improved spatial detail from the panchromatic data.

Primary advantage of pan sharpening is that it preserves the spectral characteristics of the multispectral image while incorporating the improved spatial detail from the panchromatic image. This results in a high-resolution color image with improved visual interpretation and feature discrimination.

Pan sharpening is commonly used in remote sensing applications, including land cover classification, change detection, urban planning, and natural resource management.

### Radiometric Calibration and Atmospheric Correction

Landsat8 level 1 products are delivered in 16-bit unsigned integer format (0-65535) which can be re-scaled to the Top Of Atmosphere (TOA) reflectance and/or radiance using radiometric re-scaling coefficients provided in the metadata file (MTL file). The MTL file also contains the thermal constants needed to convert TIRS data to the at satellite brightness temperature.

To implement radiometric calibration, we will complete the following processing steps using the **radCor()** function of RStoolbox:

-   Conversion of DN values to spectral radiance at the sensor

-   Conversion of spectral radiance to reflectance at sensor

-   Atmospheric correction and conversion of sensor reflectance to surface reflectance

Th following methods are available in **radCor()**:

-   rad: Radiance

-   apref: Apparent reflectance (top-of-atmosphere reflectance)

-   dos: Dark object subtratction following Chavez (1989)

-   costz: Dark object subtraction following Chavez (1996)

-   sdos: Simple dark object subtraction. Classical DOS, Lhaze must be estimated for each band separately.

First, we will read MTL file and then create a layer stack based on MLT (metadata) file:

```python
%%R
mtlFile =paste0(clipData,"LC08_L1TP_137044_20230203_20230209_02_T1_MTL.txt")
metaData= readMeta(mtlFile) # read metadata file
summary(metaData)
```

```python
%%R
lsat <- stackMeta(mtlFile)
lsat
```

### Conversion of DN values at spectral radiance at the sensor

Converting Digital Number (DN) values from Landsat imagery to spectral radiance involves a series of calculations that take into account sensor-specific calibration parameters. Spectral radiance represents the amount of electromagnetic radiation (light) measured by the sensor in each spectral band.

Landsat provides calibration parameters in the metadata file associated with each image. You can find these parameters in the Landsat Metadata File (MTL file). Look for values like the radiance gain (gain) and radiance bias (bias) for each spectral band.

For Landsat 8, a linear equation is used to convert from DN to radiance (L):

 **Radiance = (DN \* gain) + bias**

    -   **DN:** The Digital Number value for a specific pixel in the band.

    -   **gain:** The radiance gain value specific to the band.

    -   **bias:** The radiance bias value specific to the band.

We will use method = "rad" to convert DN to radiance at sensor level

```python
%%R
lsat_tra <- radCor(lsat,    # landsat stack
    metaData = metaData,      # metadata (MTL file)
    method = "rad")           # convert DN to radiance
lsat_tra
```

### Conversion of spectral radiance to reflectance at sensor

When spectral radiance values are compared within and between sensors, variations occur frequently because of difference in sun-target-sensor configurations. Hence, these data are often converted to **Top of Atmosphere (TOA)** or at sensor reflectance. It eliminates the effect of different solar zenith angles at the time of image acquisition, minimizes the solar radiance band differences occurring between different sensors, and accounts for differences in the Earth-Sun distance.

First, convert the spectral radiance values to top-of-atmosphere (TOA) reflectance using the following formula:

**TOA Reflectance = (π \* Radiance) / (ESUN \* cos(θ))**

-   **Radiance:** Spectral radiance values obtained from the previous step.

-   **π:** The mathematical constant pi (approximately 3.14159).

-   **ESUN:** Solar irradiance at the Earth's mean-sun distance for the specific band, which is also provided in the Landsat metadata.

-   **θ:** Solar zenith angle, which can be calculated based on the acquisition time and location of the image.

The  method = "apref" convert DN to apparent reflectance (top-of-atmosphere reflectance)

```python
%%R
lsat_tre<- radCor(lsat,        # landsat stack
    metaData = metaData,       # metadata
    method = "apref")          # convert DN top-of-atmosphere reflectance
lsat_tre
```

### Atmospheric correction and Conversion of sensor reflectance to surface reflectance

TOA reflectance doesn’t take atmospheric effects into account. It’s simply a relationship between the at-site radiance of the sensor and the incoming radiance of the sun, taking into account the sensor’s distance and the target’s angle. Any radiation coming out of dark pixels is assumed to be atmospheric haze, not the surface’s reflectance. The haze values are estimates of path radiance, which can be subtracted in a dark object subtraction (DOS) procedure. Since atmospheric haze affects almost everything in the visible spectrum, it’s common to only estimate haze in the blue, green, red and sometimes NIR bands.

The atmospheric correction method (solar, dos, costz) applies to the optical part of the spectrum, and doesn’t affect thermal band. If either "dos" or "costz" are selected, radCor will use the atmospheric haze decay model described by Chavez (1989). Depending on the atmosphere, the following coefficients are used:

veryClear 	λ^{-4.0}

clear 	λ^{-2.0}

moderate 	λ^{-1.0}

hazy 	λ^{-0.7}

veryHazy 	λ^{-0.5}

Dark object subtraction relies on estimating atmospheric haze by looking at dark pixels. Because dark pixels have no reflectance at all, they’re called “dark pixels”.In other words, any radiation coming out of these dark pixels is supposed to be atmospheric haze.

For Dark Object Subtraction (DOS), first we have to estimate the digital number (DN) pixel value of dark objects for the visible wavelength range using estimateHaze() function, then use these values convert sensor reflectance to surface reflectance

```python
%%R -w 450 -h 450 -u px
# Automatic haze estimation
hazeDN <- estimateHaze(lsat,       # raster object
    hazeBands = 2:5,               # haze band B2,B3, B4 and B5
    plot = TRUE,                   # display hisgram and haze values
    darkProp=0.01)
```

We will convert the DN to values to surface reflectance using using the function radCor() with estimated haze and the simple dark object subtraction, **sdos**, argument.

```python
%%R
lsat_sref <- radCor(lsat,       # raster object
    metaData = metaData,        # metadata
    method = "sdos",            # Simple dark object subtraction
    hazeValues = hazeDN,        # haze value
    hazeBands = 2:5,            # Bands corresponding to hazeValues
    atmosphere = "veryClear")    # Atmospheric characteristics
lsat_sref
```
